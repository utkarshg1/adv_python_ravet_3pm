### Trignometric functions

```python
# These angles are in degrees
ang_deg = [0, 15, 30, 45, 60, 75, 90]
```

1 degree = pi / 180 radians

```python
# Convert angles to radians
ang_rad = np.radians(ang_deg)
ang_rad
```

```python
np.sin(ang_rad)
```

```python
np.cos(ang_rad)
```

```python
np.tan(ang_rad)
```
