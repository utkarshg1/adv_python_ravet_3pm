# Assignment from OOP

```python
from dataclasses import dataclass
from math import sqrt
```

```python
@dataclass
class Triangle:
    a: int | float
    b: int | float
    c: int | float

    def __post_init__(self):
        if self.a <= 0 or self.b <= 0 or self.c <= 0:
            raise ValueError("Sides of Triangle cannot be negative ")

        if not self.is_triangle():
            raise ValueError("Sum of two sides should always be greater than third")

    def is_triangle(self) -> bool:
        return (
            (self.a + self.b > self.c)
            and (self.b + self.c > self.a)
            and (self.a + self.c > self.b)
        )

    def perimeter(self) -> int | float:
        return self.a + self.b + self.c

    def area(self) -> float:
        s = self.perimeter() / 2
        a = sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
        return a
```

# Polymorphism in class

```python
class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:

    def capital(self):
        print("Washington D.C. is capital of USA")

    def language(self):
        print("English is widely spoken in USA")
```

```python
class France:

    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
c2 = USA()
c3 = France()
```

```python
type(c1)
```

```python
type(c2)
```

```python
type(c3)
```

```python
c1.capital()
```

```python
c1.language()
```

```python
c2.capital()
```

```python
c2.language()
```

```python
for country in (c1, c2, c3):
    print(type(country))
    country.capital()
    country.language()
    print("=" * 50 + "\n")
```

### Practical example
Calulating area and perimeter for different shapes

```python
@dataclass
class Rectangle:
    length: int | float
    width: int | float

    def perimeter(self):
        return 2 * (self.length + self.width)

    def area(self):
        return self.length * self.width
```

```python
from math import pi
```

```python
pi
```

```python
@dataclass
class Circle:
    radius: int | float

    def perimeter(self):
        return 2 * pi * self.radius

    def area(self):
        return pi * (self.radius**2)
```

```python
shapes = (
    Triangle(11, 12, 13),
    Circle(7.5),
    Rectangle(50, 20),
    Triangle(10, 11, 12.5),
    Triangle(13, 14, 15),
    Circle(21),
)
```

```python
for i in shapes:
    print(i)
    print(type(i))
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"Area : {i.area():.2f}")
    print("=" * 50 + "\n")
```
